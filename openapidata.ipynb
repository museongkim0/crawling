```python
from bs4 import BeautifulSoup

html = """
<html>
    <body>
        <h1 id='title'>[1]크롤링이란?</h1>;
        <p class='cssstyle'>웹페이지에서 필요한 데이터를 추출하는 것</p>
        <p id='body' align='center'>파이썬을 중심으로 다양한 웹크롤링 기술 발달</p>
    </body>
</html>
"""

soup = BeautifulSoup(html, "html.parser")
```

```python
# 태그로 검색 방법
title_data = soup.find('h1')

print(title_data)
print(title_data.string)
print(title_data.get_text())
```

```python
# 가장 먼저 검색되는 태그를 반환
p_data = soup.find('p')

print(p_data)
print(p_data.string)
print(p_data.get_text())
```

```python
# 태그에 있는 id로 검색
title_data = soup.find(id='title')

print(title_data)
print(title_data.string)
print(title_data.get_text())
```

```python
# HTML 태그와 CSS class를 활용해 필요한 데이터 추출 방법
p_data = soup.find('p', class_='cssstyle') # 'class_='은 생략 가능

print(p_data)
print(p_data.string)
print(p_data.get_text())
```

```python
# HTML 태그와 태그에 있는 속성:속성값을 활용해 필요한 데이터 추출 방법
p_data = soup.find('p', attrs={'align':'center'})

print(p_data)
print(p_data.string)
print(p_data.get_text())
```

```python
# find_all() 관련된 모든 데이터를 리스트 형태로 추출하는 함수
p_data = soup.find_all('p')

print(p_data)
print(p_data[0].get_text())
print(p_data[1].get_text())
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://v.daum.net/v/20170615203441266')
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
# 태그 검색
soup.find('title')

# select 함수는 리스트 형태로 전체 반환
title = soup.select('title')[0]
print(title)
print(title.get_text())
```

```python
# 띄어쓰기가 있다면 하위 태그를 검색
title = soup.select('html head title')[0]
print(title.get_text())
```

```python
# 띄어쓰기는 바로 직계 자식이 아니여도 관계 없음
title = soup.select('html title')[0]
print(title.get_text())
```

```python
# >를 사용하는 경우 바로 아래의 자식만 검색
## 바로 아래 자식이 아닌 경우 에러 발생
title = soup.select('html > title')[0]
print(title.get_text())
```

```python
## 바로 아래 자식 검색
title = soup.select('head > title')[0]
print(title.get_text())
```

```python
# .은 태그의 클래스를 검색 (class가 article_view인 태그 검색)
body = soup.select('.article_view')[0]
print(type(body), len(body))
for p in body.find_all('p'):
    print(p.get_text())
```

```python
# div 태그 중 class가 article_view인 태그 검색
body = soup.select('div.article_view')[0]
print(type(body), len(body))
for p in body.find_all('p'):
    print(p.get_text())
```
